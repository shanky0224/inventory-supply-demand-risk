## 5.1 Business Insights (from EDA + Feature Importance)

### 1. Inventory Turnover & Reorder Pressure drives risk
* Items with low turnover + high reorder pressure are most likely to be high risk
* Indicates overstocking of slow-moving items

### 2. Older inventory increases mismatch probability
* Higher inventory_age_days strongly correlates with risk
* Suggests expiry loss and capital blockage

## 5.2 Actionable Recommendations

### For Operations Team
* Flag items with high reorder pressure + low turnover
* Review reorder levels every 14 days for flagged items
* Reduce reorder quantities for slow-moving SKUs

### For Inventory Managers
* Introduce age-based clearance rules
* Discount or bundle items older than threshold (e.g., 45 days)

## For Procurement
* Shift from fixed reorder rules to risk-based replenishment
* Avoid bulk ordering for items with rising mismatch risk

## 5.3 Model Usage in Real Life

### How the Model is Used
The model predicts whether a product is at high risk of supply-demand mismatch.
High-risk items are prioritized for manual review, reorder adjustment, or clearance actions.
This enables proactive inventory control and reduces waste and stockouts.
